Paper: Proprioception is All You Need: Terrain Classification for Boreal Forests


Source: https://github.com/norlab-ulaval/BorealTC?utm_source=perplexity 
Paper: https://arxiv.org/abs/2403.16877

Overall summary: Boreal forests are home to many mobility-impeding terrains that would make off-road auonomous navigation challenging. This paper introduces BorealTC, and publicly available dataset for proprioceptive-based terrain classification (TC). This data was recorded with a HuskyA200, and it contains 116 minutes of data from the following sensors: 
- inertial measurement unit (IMU)
- motor current 
- wheel odom 

It contains data from the following terrain types: 
- Snow 
- Ice 
- Silty loam (it's like soil)

The authors evaluate the following models: 
- CNN 
- State space model Mamba architecture 

Key results: 
- CNN outperforms Mamba on each separate dataset, Mamba achieves better accuracy in combined daaset 
- Mamba's learning capacity is greater for larger amounts of data
